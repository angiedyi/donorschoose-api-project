```python
#Import appropriate Python Libraries
import pandas as pd
import numpy as np
import os
import sqlalchemy
from sqlalchemy import create_engine, MetaData
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Numeric, Text, Float, Date, String
from sqlalchemy import create_engine
import psycopg2
```

```python
donorschoose_data = 'donorschoose_api.csv'
```

```python
donorschoose_df = pd.read_csv('donorschoose_api.csv')
```

```python
donorschoose_df
```

```python
#Create an engine to a postgressql datbase file
engine = create_engine('postgresql+psycopg2://angieyi@localhost/donorschooseapi')
```

```python
#Create a connection to the engine called `conn`
conn = engine.connect()
```

```python
# Use `declarative_base` from SQLAlchemy to connect your class to your database
Base = declarative_base()
```

```python
#Create a Class for Projects

class Projects(Base):
    __tablename__ = 'projects'
    
    
    id = Column(Integer, primary_key=True)
    cityname = Column(String)
    costtocomplete = Column(Integer)
    low_poverty = Column(Integer)
    moderatelow_poverty = Column(Integer)
    moderate_poverty = Column(Integer)
    moderatehigh_poverty = Column(Integer)
    high_poverty = Column(Integer)
    highest_poverty = Column(Integer)
    unknown_poverty = Column(Integer)
    total_projects = Column(Integer)
    avg_household_salary = Column(Integer)
    
    def __repr__(self):
        return f"id={self.id}, name={self.name}"
```

```python
# Use `create_all` to create the table in the database
Base.metadata.create_all(engine)
```

```python
projects_data = donorschoose_df.to_dict(orient='records')
```

```python
metadata = MetaData(bind=engine)
metadata.reflect()
```

```python
projects_table = sqlalchemy.Table('projects', metadata, autoload=True)
```

```python
conn.execute(projects_table.insert(), projects_data)
```
